```python
# importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data for input features and output feature
# initialize or take input from user
X = np.array([
    1.2, 2.3, 3.1, 4.8, 5.5, 6.7, 7.2, 8.1, 9.0, 10.4,
    11.3, 12.5, 13.7, 14.4, 15.8, 16.3, 17.5, 18.9, 19.4, 20.8,
    21.3, 22.5, 23.6, 24.1, 25.7, 26.2, 27.8, 28.4, 29.1, 30.5,
    31.2, 32.7, 33.3, 34.8, 35.6, 36.2, 37.1, 38.5, 39.7, 40.8,
    41.5, 42.9, 44.1, 45.3, 46.6, 47.8, 48.2, 49.4, 50.7, 51.0
])

Y = np.array([
    3.4, 4.5, 5.9, 7.2, 8.1, 9.8, 10.2, 11.3, 12.4, 13.7,
    14.1, 15.6, 16.2, 17.9, 18.4, 19.2, 20.1, 21.3, 22.6, 23.2,
    24.7, 25.4, 26.8, 27.9, 28.3, 29.1, 30.5, 31.7, 32.3, 33.6,
    34.9, 35.4, 36.1, 37.8, 38.4, 39.9, 40.2, 41.6, 42.3, 43.7,
    44.9, 45.5, 46.7, 47.2, 48.9, 49.6, 50.8, 51.3, 52.7, 53.9
])
#X = np.array([[1, 4], [2, 5], [3, 8], [4, 2]]) # for multiple linear regression
#Y = np.array([1, 6, 8, 12]) # for multiple linear regression
```

```python
# plotting data points
plt.scatter(X, Y, color='blue')
plt.title('Scatter Plot of X vs. Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
# splitting into train and test data
Xtrain = X[:40]
Ytrain = Y[:40]
Xtest = X[40:]
Ytest = Y[40:]
X = Xtrain
Y = Ytrain
```

```python
# initialize X0, theta vector
X0 = np.ones(len(Y))
theta = np.zeros(2)
#theta = np.zeros(len(X[0])+1) # for multiple linear regression
```

```python
# tune learning rate and difference in loss function for termination 
learningRate = 1e-5
c = 1e-5
```

```python
# gradient descent algorithm
Ltheta = 0
while (True): 
    Ltheta_old = Ltheta
    Ltheta = 0
    gradient = np.zeros(len(theta))
    for i in range(len(Y)): 
        Xi = np.array([X0[i], X[i]]) # computing Xi vector
        #Xi = np.array([X0[i]] + [j for j in X[i]]) # for multiple linear regression
        Ltheta += (Y[i] - np.dot(theta, Xi))**2 # computing Ltheta
        gradient += (-2 * (Y[i] - np.dot(theta, Xi)) * Xi) # computing gradient 
    theta = theta - (learningRate * gradient) # computing new theta vector
    print(f"Ltheta: {Ltheta}")
    if(abs(Ltheta_old - Ltheta) < c):
        break
print(theta)
```

```python
# making prediction on test data
Ypred = []
for i in Xtest:
    y = theta[0] + theta[1] * i
    Ypred.append(y)
Ypred = np.array(Ypred)
print(Ypred)
```

```python
# computing RMSE
rmse = np.sqrt(np.mean((Ytest - Ypred) ** 2))
print(rmse)
```

```python
# plotting actual vs predicted
plt.figure(figsize=(10, 6))
plt.plot(Xtest, Ytest, color='blue', label='Actual Values')
plt.plot(Xtest, Ypred, color='red', label='Predicted Values')
plt.title('Actual vs. Predicted Values')
plt.legend()
plt.grid(True)
plt.show()
```
